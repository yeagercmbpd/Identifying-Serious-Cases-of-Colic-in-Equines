## Download and pepare horse colic dataset from UCI repository for Logistic Calssification model

```python
#Load all necessary libraries and packages.
import pandas as pd
import numpy as np
import re
import string
import itertools

#https://archive.ics.uci.edu/ml/machine-learning-databases/horse-colic/horse-colic.data
```

The data is provided in 2 files, they split it to test (300 rows) and train (68 rows) sets on their own. I've opted to combine these two into a single dataframe and perform my own test train split. The two cells below download and load the two into pandas df's and append them together.


```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/horse-colic/horse-colic.data',sep='\ ', header=None)
df
```

```python
df2=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/horse-colic/horse-colic.test',sep='\ ', header=None)
df2
```

```python
df=df.append(df2)
```

The data dictionary for this set, which is totally encoded, is provided in a secondary text file. The below cells read this text file and and selects those lines from it which include the feature attirbute descriptions for further consideration.

```python
cols = open('HorseColic/horse-colic.names', "r")
linelist=cols.readlines()
```

```python
matching = [s for s in linelist if "Attribute" in s]
X=" ".join(matching)
```

```python
MyAttributes=linelist[linelist.index(X):]
MyAttributes
```

```python
#The following columns are not needed per the documentation and are dropped.
df=df.drop([27,26,25],axis=1)
```

While the needed lines are selected and passed into list format above, the follwing re code locates the feature names specificially and reads them into a list, so that the columns within my dataframe can be named for easier analysis within the model.

```python
colnames=[]
for i in range(len(MyAttributes)):
    colnames.append(re.findall("\d\:(.*)",MyAttributes[i]))
```

```python
colnames=list(filter(None, colnames))
```

```python
#Remove any leftover punctuation using list proided by string
for d in range (len(colnames)):
    colnames[d] = [''.join(c for c in s if c not in string.punctuation) for s in colnames[d]]
```

```python
columns=(list(itertools.chain.from_iterable(colnames)))
columns
```

```python
#Remove any leftover whitespace after punctuation is removed
columnnames=[]
for i in range(len(columns)):
    columnnames.append(str(columns[i]).strip())
columnnames
```

```python
#Drop last label as its correspondinf column was dropped above
columnnames=(columnnames[:-1])
```

```python
len(columnnames)
```

```python
df.columns=columnnames
df._data = df._data.rename_axis(lambda x: x, 0, True)
```

I have decided to drop any columns which have a null value proportion of greater than 25%

```python
df1 = df.apply(pd.to_numeric, args=('coerce',))

null_counts=round(df1.isnull().sum()/len(df) *100,2)
remove=null_counts[null_counts > 25].index
```

```python
null_counts
```

```python
remove
```

```python
df1=df1.drop(list(remove), axis=1)
df1=df1.drop('Hospital Number', axis=1)
```

Final df is named simple binary, outcome is my first target variable as whether the horse lives or dies, according to the documentation 2 is death and 3 is euthenized. I've decided to comine these into one so that a binary classification can be made using out model of lvives or dies. This recoded varibale is written to the new column Target


```python
SimpleBinary=df1
SimpleBinary['outcome'].value_counts()
```

```python
#SimpleBinary=SimpleBinary[SimpleBinary['outcome'] != '?']
#SimpleBinary.replace('?',0,inplace=True)
```

```python
SimpleBinary.shape
```

```python
SimpleBinary['Target'] = SimpleBinary['outcome'].map(lambda x: 2 if x == 3 else x)

SimpleBinary=SimpleBinary.astype(float)
SimpleBinary['Target'].value_counts()
```

```python
#originally encoded 1 or 2, i prefer 0 or 1, recoded here.
SimpleBinary['Target'].replace(to_replace =1, value =0,inplace=True)
SimpleBinary['Target'].replace(to_replace =2, value =1,inplace=True)
```

```python
SimpleBinary['Target'].value_counts()
```

```python
SimpleBinary.isna().sum()
```

Some variables are categorical some continuous. The below features are described within the documentation as continuous. Empty values for these are extrapolated using the median values throughout the dataset.

```python
LinearNumVals=['rectal temperature','pulse','respiratory rate','packed cell volume','total protein']
```

```python
for colz in SimpleBinary.columns.values:
    SimpleBinary[LinearNumVals]=SimpleBinary[LinearNumVals].fillna(SimpleBinary[LinearNumVals].median())
```

Splitting/decoding type of leison column into location and type. The documetation states that the first digit of this numeric value denotes location and the second digit denotes type of lesion. null is used for a lack of reported information.


```python
SimpleBinary['location of lesion']=SimpleBinary['type of lesion'].astype(str).str[0:1]

SimpleBinary['type of lesion']=SimpleBinary['type of lesion'].astype(str).str[1:2]

SimpleBinary['type of lesion'].replace(to_replace ='.', value =0,inplace=True)
```

Though protien level is continuous value its not simply ordinal. per documentation, recode range as normal '0' or abnomral '1'

```python
SimpleBinary['total protein'] = np.where(SimpleBinary['total protein'] > 7.5, 1, SimpleBinary['total protein'])
SimpleBinary['total protein'] = np.where(SimpleBinary['total protein'] < 6, 1, SimpleBinary['total protein'])
SimpleBinary['total protein'] = np.where(SimpleBinary['total protein'].between(6,7.5), 0, SimpleBinary['total protein'])
```

```python
#Any leftover na values for categorical varibales after this cleanup are filled with 0
SimpleBinary=SimpleBinary.fillna(0) 
```
